```python
# Import BeautifulSoup and splinter
from bs4 import BeautifulSoup as soup
from splinter import Browser
import pandas as pd
import numpy as np
import csv
```

```python
# Set up Splinter
browser = Browser('chrome')
```

```python
# Visit the Travellers Choice Awards - Beaches of Trip Advisor website
url = 'https://www.tripadvisor.com/TravelersChoice-Destinations'
browser.visit(url)
# Optional delay for loading the page
browser.is_element_present_by_css('div.list_text', wait_time=1)
```

```python
# Scrape the website html
html = browser.html

# Create a BeautifulSoup object from the scraped HTML
beaches = soup(html, 'html.parser')
```

### Rank and Country Name of Beach

```python
# Get Rank and name of beach
names = beaches.find_all(class_="mainName")
namelist = []
for i in names:
    href=i.find("a")    
    text_data=href.text       
    namelist.append(text_data.split('. '))
    
print(namelist)
```

```python
# Get image urls
images = beaches.find_all(class_="sizedThumb_container")
iurls = []
for i in images:
    href=i.find("img")
    iurls.append(href["src"])
    
print(iurls)
```
